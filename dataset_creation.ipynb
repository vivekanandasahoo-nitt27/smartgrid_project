```python
import pandas as pd

```

```python
# Load datasets
healthy = pd.read_csv("healthy_dataset.csv")
voltage_attack = pd.read_csv("voltage_attack_dataset.csv")
attack = pd.read_csv("attack_dataset.csv")

# Concatenate all datasets
final_dataset = pd.concat(
    [healthy, voltage_attack, attack],
    ignore_index=True
)

# Shuffle dataset
final_dataset = final_dataset.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)

# Save final dataset
final_dataset.to_csv("final_dataset.csv", index=False)

# Check result
print("Final Shape:", final_dataset.shape)

print("\nClass Distribution:")
print(final_dataset['class'].value_counts())
```
